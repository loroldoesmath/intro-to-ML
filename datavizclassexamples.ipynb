```python
## BAR CHART EXAMPLE

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import seaborn as sns

# sample data
shoe_names = ["Nike Air Max", "Nike Air Force 1", "Nike Dunk Low", "Nike Blazer Mid",
    "Nike Pegasus", "Nike React Infinity Run", "Nike ZoomX Vaporfly",
    "Nike Air Zoom", "Nike Free RN", "Nike Metcon"]

np.random.seed(42)  # For reproducibility
sales_quantities = np.random.randint(100, 1000, len(shoe_names))

# Create a DataFrame
data = pd.DataFrame({
    "Shoe Name": shoe_names,
    "Sales Quantity": sales_quantities
})

num_colors = len(data["Shoe Name"])

# Get a colormap
colormap = cm.get_cmap('tab10', num_colors)  # 'tab10' is a good colormap for categorical data

# Create a bar plot with different colors for each shoe using the colormap
plt.figure(figsize=(10, 6))
bars = plt.bar(data["Shoe Name"], data["Sales Quantity"], color=[colormap(i) for i in range(num_colors)])
plt.xlabel("Nike Shoe Models")
plt.ylabel("Sales Quantity")
plt.title("Sales Quantities of Nike Shoes")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
## HEAT MAP EXAMPLE

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import seaborn as sns

#sample data: rows are flavors, columns are days of the week
data = np.array([
    [120, 130, 115, 140, 150, 170, 160],  # Chocolate
    [80, 85, 78, 90, 100, 110, 105],      # Vanilla
    [60, 65, 55, 70, 75, 80, 85],         # Strawberry
    [45, 40, 50, 55, 60, 65, 50],         # Mint
    [70, 75, 80, 90, 85, 95, 100],        # Cookie Dough
])

#labels
flavors = ["Chocolate", "Vanilla", "Strawberry", "Mint", "Cookie Dough"]
days = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]

#plotting the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(data, annot=True, cmap="coolwarm", xticklabels=days, yticklabels=flavors)
plt.title("Ice Cream Flavor Popularity Heatmap")
plt.xlabel("Days of the Week")
plt.ylabel("Flavors")
plt.show()
```

```python
import plotly.graph_objects as go
import pandas as pd
import numpy as np

# Sample data
np.random.seed(0)
data = pd.DataFrame({
    'x': np.random.rand(100),
    'y': np.random.rand(100),
    'category': np.random.choice(['A','B','C'], 100)
})

# Scatter plot
fig = go.Figure()

# Scatter traces
for category in data['category'].unique():
    subset = data[data['category'] == category]
    fig.add_trace(go.Scatter(
        x=subset['x'],
        y=subset['y'],
        mode='markers',
        marker=dict(size=10, opacity=0.7),
        name=category
    ))

# Add labels
fig.update_layout(
    title='LW Interactive Scatter Plot',
    xaxis_title='X-axis',
    yaxis_title='Y-axis',
    legend_title = 'Category',
    template = 'plotly_dark'
)

fig.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from math import pi

#defining the data
categories = ['Speed', 'Strength', 'Agility']
athletes = {
    'Athlete 1': [90, 85, 80],
    'Athlete 2': [70, 95, 85],
    'Athlete 3': [85, 80, 90]
}

#number of variables
num_vars = len(categories)

#compute angle for each category
angles = [n / float(num_vars) * 2 * pi for n in range(num_vars)]
angles += angles[:1]

#initialize the radar chart
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))

#draw one athlete at a time
for athlete, scores in athletes.items():
    # Close the loop
    scores += scores[:1]
    ax.plot(angles, scores, linewidth=2, linestyle='solid', label=athlete)
    ax.fill(angles, scores, alpha=0.25)

#add category labels
plt.xticks(angles[:-1], categories)

#draw ylabels
ax.set_rlabel_position(30)
plt.yticks([20, 40, 60, 80, 100], ["20", "40", "60", "80", "100"], color="grey", size=7)
plt.ylim(0, 100)

#add legend
plt.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))

#display radar chart
plt.show()
```

```python
## GANTT CHART EXAMPLE

import plotly.express as px
import pandas as pd

#sample data
df = pd.DataFrame([
    dict(Task="Foundation", Start='2009-01-01', Finish='2009-02-28'),
    dict(Task="Framing", Start='2009-03-05', Finish='2009-04-15'),
    dict(Task="Roofing", Start='2009-02-20', Finish='2009-05-30')
])

fig = px.timeline(df, x_start="Start", x_end="Finish", y="Task")
fig.update_yaxes(autorange="reversed") # otherwise tasks are listed from the bottom up
fig.show()
```

```python
## PARETO CHART EXAMPLE
import pandas as pd
import matplotlib.pyplot as plt

# sample data
data = {
    'Complaint Type': [
        'Slow Service',
        'Cold Food',
        'Wrong Order',
        'Unfriendly Staff',
        'Noisy Environment',
        'High Prices',
        'Uncleanliness',
        'Limited Menu',
        'Parking Issues'
    ],
    'Frequency': [150, 120, 90, 80, 70, 60, 50, 40, 30]
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Step 2: Sort the data in descending order
df = df.sort_values(by='Frequency', ascending=False)

# Step 3: Calculate cumulative percentage
df['Cumulative Percentage'] = df['Frequency'].cumsum() / df['Frequency'].sum() * 100

# Step 4: Plot the Pareto Chart
fig, ax = plt.subplots(figsize=(16,6))

# Plot bar chart for frequencies
ax.bar(df['Complaint Type'], df['Frequency'], color='skyblue')

# Plot line chart for cumulative percentage
ax2 = ax.twinx()
ax2.plot(df['Complaint Type'], df['Cumulative Percentage'], color='orange', marker='o', linestyle='-')
ax2.set_ylim(0, 100)

# Add labels and title
ax.set_xlabel('Complaint Type', fontdict=dict(weight='bold'))
ax.set_ylabel('Frequency', fontdict=dict(weight='bold'))
ax2.set_ylabel('Cumulative Percentage (%)', fontdict=dict(weight='bold'))
plt.title('Pareto Chart of Customer Complaints in a Restaurant', fontdict=dict(weight='bold'))

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right', )
ax.tick_params(axis='x', labelsize=10)
plt.subplots_adjust(bottom=0.1)
# Show grid for the secondary y-axis
ax2.yaxis.grid(True)

# Display the chart
plt.tight_layout()
plt.show()
```
